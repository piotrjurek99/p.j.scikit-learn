```python
import pandas as pd

path = "/content/drive/MyDrive/Projekt zaliczeniowy/diabetes_012_health_indicators_BRFSS2015.csv"
df = pd.read_csv(path)
```

2. Weryfikacja ilości wartości na daną klasę. Można zauważyć dużą dysproporcję każdej z nich

```python
df['Diabetes_012'].value_counts()
```

```python
df1 = df.loc[df['Diabetes_012'] == 0.0]
```

```python
df1
```

3. Redukcja ilości wierszy należących do danej klasy, aż do poziomu, w którym każda z nich posiadała taką samą ilość sampli - 4500. Te same metody zastosowano dla każdej z klas. 

```python
N = 249180
df1 = df1.tail(df.shape[0] -N)

df1
```

```python
df1['Diabetes_012'].value_counts()
```

```python
df2 = df.loc[df['Diabetes_012'] == 2.0]
df2
```

```python
M = 30868
df2 = df2.tail(df.shape[1] -M)
df2
```

```python
df2['Diabetes_012'].value_counts()
```

```python
df3 = df.loc[df['Diabetes_012'] == 1.0]
df3
```

```python
O = 153
df3 = df3.tail(df.shape[1] -O)
df3
```

```python
df3['Diabetes_012'].value_counts()
```

4. Połączenie 3 zredukowanych Dataframów każdej z klasy do wspólnego Datasetu (df4) 

```python
dfs = [df1, df2, df3]
df4 = pd.concat([df.squeeze() for df in dfs], ignore_index=True)

df4
```

5. Uzyskanie równej ilości sampli dla każdej z klas

```python
df4['Diabetes_012'].value_counts()
```

6. Randomizacja sampli

```python
from sklearn.utils import shuffle
df4 = shuffle(df4)
```

```python
df4
```
