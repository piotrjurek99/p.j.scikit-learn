5. Tworzenie pierwszego modelu z biblioteką scikit-learn - DecisionTree.
Ustalenie metryk accuracy oraz roc_auc, które będą obliczne, podział danych na testowe i treningowe.

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score

clf = DecisionTreeClassifier()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
clf.fit(X_train, Y_train)
```

```python
def validateModel(x, y, clf):
  y_score = clf.predict_proba(x)
  y_pred = clf.predict(x)
  acc = accuracy_score(y, y_pred)
  roc = roc_auc_score(y, y_score, multi_class='ovr')
  return (acc, roc)
```

```python
validateModel(X_test, Y_test, clf)
```

```python
validateModel(X_train, Y_train, clf)
```

```python
from sklearn.tree import DecisionTreeClassifier

def trainModel(max_depth, X_train, Y_train, X_test, Y_test):
  
  clf = DecisionTreeClassifier(max_depth=max_depth)
  clf.fit(X_train, Y_train)
  perf_train = validateModel(X_train, Y_train, clf)
  perf_test = validateModel(X_test, Y_test, clf)

  return perf_train, perf_test, clf
```

```python
from tqdm import tqdm

results = []

for d in tqdm(range(1,11)):
  results.append((d, trainModel(d, X_train, Y_train.values.ravel(), X_test, Y_test)))
```

6. Wizualizacja wyników modelu DecisionTree

```python
import matplotlib.pyplot as plt

plt.plot([x[0] for x in results], [x[1][0][1] for x in results], label="train")
plt.plot([x[0] for x in results], [x[1][1][1] for x in results], label="test")
plt.legend()
plt.show()
```

```python
results
```

7. Określenie najlepszego przypadku z 10 modeli - max_depth=10

```python
best_model = results[9][1][2]
best_model
```

8. Wizualizacja confusion matrix dla algorytmu DecisionTree

```python
from sklearn.metrics import ConfusionMatrixDisplay

y_pred = best_model.predict(X)

ConfusionMatrixDisplay.from_predictions(Y, y_pred, normalize='pred')
plt.show()
```

9. Analogiczne przygotowanie drugiego modelu scikit-learn - GradientBoost

```python
from sklearn.ensemble import GradientBoostingClassifier

gboost = GradientBoostingClassifier()

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
gboost.fit(X_train, Y_train.values.ravel())

def validateModel(x, y, gboost):
  y_score = gboost.predict_proba(x)
  y_pred = gboost.predict(x)
  acc = accuracy_score(y, y_pred)
  roc = roc_auc_score(y, y_score, multi_class='ovr')
  return (acc, roc)
```

```python
validateModel(X_test, Y_test, gboost)
```

```python
validateModel(X_train, Y_train, gboost)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

def trainModel(max_depth, X_train, Y_train, X_test, Y_test):
  
  gboost = GradientBoostingClassifier(max_depth=max_depth)
  gboost.fit(X_train, Y_train)
  perf_train = validateModel(X_train, Y_train, gboost)
  perf_test = validateModel(X_test, Y_test, gboost)

  return perf_train, perf_test, gboost
```

```python
from tqdm import tqdm

results = []

for d in tqdm(range(1,11)):
  results.append((d, trainModel(d, X_train, Y_train.values.ravel(), X_test, Y_test)))
```

10. Wizualizacja wyników modelu GradientBoost

```python
import matplotlib.pyplot as plt

plt.plot([x[0] for x in results], [x[1][0][1] for x in results], label="train")
plt.plot([x[0] for x in results], [x[1][1][1] for x in results], label="test")
plt.legend()
plt.show()
```
